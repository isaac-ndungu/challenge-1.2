## Exercise 2

```python
# print('Gross pay Calculator')

hours_input = input('Enter hours:')
hours = int(hours_input)
print(f'Hours: {hours}')

rate_input = input('Enter rate per hour')
rate = float(rate_input)
print(f'Rate: {rate}' )

pay = hours * rate
print(f'Pay: {pay}')
```

## Exercise 3

```python
width = 17
height = 12.0

floor_division = width // 2
print(f'width // 2 = {floor_division} \n {type(floor_division)}\n')

division = width / 2.0
print(f'width / 2.0 = {division} \n {type(division)}\n')

division_h = height / 3
print(f'height / 3 = {division_h} \n {type(division_h)}\n')

total = 1 + 2 * 5
print(f'1 + 2 * 5 = {total} \n {type(total)}')

```
